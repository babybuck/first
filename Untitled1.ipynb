```python
import numpy as np
import pandas as pd
```

```python
df_topic = pd.read_csv("Topic_Survey_Assignment.csv", index_col = 0)
df_topic
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
topic = df_topic.sort_values(by='Very interested', ascending=False)
topic
```

```python
topic_pct = topic.div(2233).multiply(100).round(2)
topic_pct
```

```python
colors = {"Very interested":"#5cb85c", "Somewhat interested":"#5bc0de", "Not interested":"#d9534f"}

ax =topic_pct.plot(kind='bar', figsize=(20, 8), color=colors, fontsize=(14), width=(.8))

plt.title('Percentage of Respondents Interest in Data Science Areas', fontsize=(16))
plt.legend(fontsize=(14))

plt.tick_params(axis='y', which='both', left=False, right=False, labelleft=False)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

for p in ax.patches:
    ax.annotate("%.2f" % p.get_height(), 
    (p.get_x() + p.get_width()/2., p.get_height()), ha='center', 
    va='center', xytext=(0,10), textcoords='offset points', fontsize=(14))

plt.show()
```

```python
df_crime= pd.read_csv("Police_Department_Incidents_-_Previous_Year__2016_.csv")
df_crime.drop(['IncidntNum', 'Category', 'Address', 'X', 'Y', 'Location', 'Descript', 'DayOfWeek', 'Date', 'Time', 'Resolution', 'PdId'], axis=1, inplace=True)
df_crime.rename(columns={'PdDistrict':'Neighborhood'}, inplace=True)

df_crime['Count'] = df_crime.sum(axis=1)
df_crime.groupby(['Neighborhood']).count()
```
